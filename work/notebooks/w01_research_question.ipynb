## 2. The question: decision, action, cost of a wrong call
Decision: Which observable signals (word count, freshness, position, CTR, engagement) should a content strategist actually pay attention to when triaging pages for review.
Who acts on it: A content strategist/SEO reviewer deciding where to spend limited review time.
Cost of a wrong call: If I claim a signal matters when it doesn't, reviewers waste time chasing it. If I miss a signal that does matter, real opportunities go unreviewed. Since this is decision-support (not automation), the cost is wasted attention — not a broken pipeline.
Unit of analysis: One row = one pseudonymized content page, measured over a trailing 90-day window.
Why data/ML helps here, not just a rule: A simple rule like "flag pages ranked below position 10" ignores that signals interact — e.g. position alone only weakly predicts CTR (r = -0.08 in this data), so a single-factor rule would misjudge most pages. With 30,000 pages and 44 columns, there are too many combinations to eyeball by hand, but this is still signal analysis, not full prediction — I'm looking for which combinations of signals are worth a reviewer's attention, not building an automated decision-maker.
